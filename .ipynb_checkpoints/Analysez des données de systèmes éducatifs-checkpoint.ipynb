Les colonnes représentant des années correspondent aux valeurs de l’indicateur pour chaque année. Les années allant de 1970 à 2017 représentent des données historiques observées, tandis que les années postérieures (2020 à 2100) correspondent à des projections ou estimations futures, et non à des données réellement observées.

Interpréter la présence d’années futures

Les données jusqu’en 2017 sont des observations réelles. Les valeurs à partir de 2020 sont des projections futures, utilisées pour l’anticipation et la planification, et ne doivent pas être interprétées comme des données observées.

Filtrer les années en conséquence

```python
# Définition des années observées (données réelles)
annees_observees = np.arange(1970, 2018).astype(str).tolist()

# Colonnes à conserver : informations d'identité + années observées
colonnes_a_conserver = [
    'CountryName',
    'CountryCode',
    'Indicator Name',
    'IndicatorCode'
] + annees_observees

# Filtrage du DataFrame (sélection explicite des colonnes)
Data_df = Data_df[colonnes_a_conserver]

# Vérifications
print("Nombre de colonnes après filtrage :", Data_df.shape[1])
print("Période conservée :", annees_observees[0], "→", annees_observees[-1])
print("Colonnes après filtrage :", Data_df.columns.tolist())

# Sauvegarde du DataFrame filtré
Data_df.to_csv("EdStatsData_v6.csv", index=False)
```

**calculez la proportion d’indicateurs**

```python
# Liste des colonnes correspondant aux années
annees = [str(year) for year in range(1970, 2017)]
```

```python
proportion_annees_par_indicateur = (
    Data_df
        .groupby('IndicatorCode')[annees]
        .apply(lambda x: x.notna().any(axis=0).mean())
        .reset_index(name='ProportionAnneesRenseignees')
)

print("\nProportion d’années renseignées par indicateur :")
print(
    proportion_annees_par_indicateur
        .sort_values(by='ProportionAnneesRenseignees', ascending=False)
)
```

**proportion d’indicateurs avec des valeurs renseignées par année**

```python
# Colonnes années
annees = [str(year) for year in range(1970, 2017)]

# S'assurer que les valeurs sont numériques
Data_df[annees] = Data_df[annees].apply(pd.to_numeric, errors='coerce')

# Proportion d'indicateurs renseignés par année
proportion_indicateurs_par_annee = Data_df[annees].notna().sum() / len(Data_df)
print("Proportion d'indicateurs renseignés par année :")
print(proportion_indicateurs_par_annee)
```

 #### réduisez le nombre d’années et d’indicateurs pour garder celles et ceux qui sont les plus riches en données

```python
# --- Seuils ---
seuil_indicateurs = 0.5
seuil_annees = 0.05

# --- 1. Filtrer les indicateurs ---
indicateurs_a_garder = proportion_annees_par_indicateur.loc[
    proportion_annees_par_indicateur['ProportionAnneesRenseignees'] >= seuil_indicateurs,
    'IndicatorCode'
].tolist()

# --- 2. Filtrer les années ---
annees_a_garder = proportion_indicateurs_par_annee.loc[
    proportion_indicateurs_par_annee >= seuil_annees
].index.tolist()

# --- 3. S'assurer que les années existent dans le DataFrame ---
annees_a_garder = [annee for annee in annees_a_garder if annee in Data_df.columns]

# --- 4. Colonnes finales à garder ---
colonnes_finales = ['Indicator Name', 'IndicatorCode'] + annees_a_garder

# --- 5. Filtrer le DataFrame principal ---
Data_df_filtre = Data_df.loc[
    Data_df['IndicatorCode'].isin(indicateurs_a_garder),
    colonnes_finales
]
```

```python
nb_pays_par_indicateur_annee = (
    Data_df.groupby('IndicatorCode')[annees_a_garder]
    .apply(lambda x: x.notna().sum())
    .reset_index()
    .melt(id_vars='IndicatorCode', var_name='Annee', value_name='NbPaysRenseignes')
)
```

```python
indicateurs_riches = (
    nb_pays_par_indicateur_annee
        .groupby('IndicatorCode')['NbPaysRenseignes']
        .mean()
        .reset_index()
)

indicateurs_riches = indicateurs_riches.sort_values(
    by='NbPaysRenseignes', ascending=False
)

top15_indicateurs = indicateurs_riches.head(15)
print(top15_indicateurs)
```
